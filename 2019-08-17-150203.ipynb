```python
y = data.Churn
```

```python
X = data[['CustServ Calls', 'Account Length', 'Day Mins']]
```

```python
X.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
t = DecisionTreeClassifier()
t.fit(X,y)
```

```python
# [CustServCalls, Account Length, Day Mins]
t.predict([[1,90,100]])
```

```python
t.predict([[4,25,100]])
```

```python
# tells how important each thing is from decision tree
t.feature_importances_
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, t.predict(X))
```

```python
# random state chooses test items randomly rather than sequentially (not first 10 items etc)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=.25, random_state=0)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
t_r = DecisionTreeClassifier()
t_r.fit(x_train, y_train)
```

```python
t_r.predict(x_test)
```

```python
accuracy_score(y_test, t_r.predict(x_test))
```
